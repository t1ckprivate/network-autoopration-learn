```python
from nornir import InitNornir
from nornir_netmiko.tasks import netmiko_send_command
from nornir_utils.plugins.functions import print_result

nr = InitNornir(config_file="hosts.yaml")

def nornir_basic_filtering(task):
    task.run(task=netmiko_send_command, command_string="display ip int brief")

nr_filter = nr.filter(type="switch", city="shenzhen")
results=nr_filter.run(task=nornir_basic_filtering)
print_result(results)
```

```python
from nornir import InitNornir
from nornir_netmiko.tasks import netmiko_send_command
from nornir_utils.plugins.functions import print_result
from nornir.core.filter import F

nr = InitNornir(config_file="hosts.yaml")

def nornir_advanced_filtering(task):
    task.run(task=netmiko_send_command, command_string="display ospf routing")

nr_filter = nr.filter( F(type="router") & F(city="guangzhou") & F(core="True") )
results=nr_filter.run(task=nornir_advanced_filtering)
print_result(results)
```
